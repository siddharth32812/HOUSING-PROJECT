```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=455,test_size=.30)
```

```python
maxsc=0
maxrs=0
for i in range(1,500):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=i)
    lm=LinearRegression()
    lm.fit(x_train,y_train)
    predlm=lm.predict(x_test)
    score=lm.score(x_train,y_train)
    if score>maxsc:
        maxsc=score
        maxrs=i
print(maxsc)
print(maxrs)
```

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
predlm=lm.predict(x_test)
print('mean absolute error:',mean_absolute_error(y_test,predlm))
print('mean squared error:',mean_squared_error(y_test,predlm))
print(lm.score(x_train,y_train))
print(r2_score(y_test,predlm))
```

```python
ls=Lasso(alpha=.0001)
ls.fit(x_train,y_train)
print(ls.score(x_train,y_train))
print(ls.coef_)
```

```python
plt.figure(figsize=(7,5))
plt.bar(x.columns,ls.coef_)
```

```python
rg=Ridge(alpha=.0001)
rg.fit(x_train,y_train)
print(rg.score(x_train,y_train))
print(rg.coef_)
```

```python
plt.figure(figsize=(7,5))
plt.bar(x.columns,rg.coef_)
```

```python
kernel_list=['linear','poly','rbf']
for i in kernel_list:
    svr=SVR(kernel=i)
    svr.fit(x_train,y_train)
    print(svr.score(x_train,y_train))
```

```python
rfr=RandomForestRegressor()
rfr.fit(x_train,y_train)
predrfr=rfr.predict(x_test)
print('predicted house_price',predrfr)
print('actual house_price',y_test)
print('score',rfr.score(x_train,y_train))
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x=y_test,y=predrfr,color='r')
plt.plot(y_test,y_test,color='b')
plt.xlabel('actual house_price',fontsize=14)
plt.ylabel('predicted house_price',fontsize=14)
plt.title('random forest regressor',fontsize=18)
plt.show()
```
